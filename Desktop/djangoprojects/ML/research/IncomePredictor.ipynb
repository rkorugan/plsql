```python
import json
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
import joblib
```

```python
df=pd.read_csv(r'C:\Users\Ravi Kumar\Desktop\DjangoProjects\ML\datasets-for-start-master\adult\data.csv')
df.head()
```

```python
df.columns
```

```python
cols=[c for c in df.columns if c != ' income']
```

```python
cols
```

```python
x=df[cols]
y=df[' income']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1234)
```

```python
x_train.isnull().sum()
```

```python
x_train.mode()
```

```python
train_mode=dict(x_train.mode().iloc[0])
```

```python
print(train_mode)
```

```python
x_train=x_train.fillna(train_mode)
```

```python
print(train_mode)
```

```python
x_train.columns=x_train.columns.str.lstrip()
```

```python
x_train.columns
```

```python
encoders={}
for column in ['workclass','education','marital-status','occupation','relationship','race','sex','native-country']:
    encdr=LabelEncoder()
    x_train[column]=encdr.fit_transform(x_train[column])
    encoders[column]=encdr
x_train  
```

```python
train_mode
```

```python
encoders
```

```python
rf=RandomForestClassifier(n_estimators = 100)
rf=rf.fit(x_train,y_train)
```

```python
et=ExtraTreesClassifier(n_estimators = 100)
et=et.fit(x_train,y_train)
```

```python
et
```

```python
rf
```

```python
joblib.dump(train_mode,"./train_mode.joblib",compress=True)
```
